### Second variable: <a name="2ndvariable"></a>
### Role models

Second variable will be if the girls have role models involved in sport.   

There is a lack of positive role models. Today’s girls are bombarded with images of external beauty, not those of confident, strong female athletic role models. To some girls, fitting within the mold that they are constantly told to stay in is more important than standing out. Peer pressure can be hard for girls at any age; when that pressure isn’t offset with strong encouragement to participate in sports and healthy physical activity, the results may lead girls to drop out altogether.[6]

The more role models girls have will encourage more participation in sport. According to published 2020 Rally Report (by Canadian Women & Sport)[7] having same-gender role models and active parents have a positive influence on girls and teenagers.   
Girls who have these kinds of role models to look up to are more likely to participate in sports and experience the many benefits, such as increased physical health and development of leadership and socialization skills.[8]

Statistics claim that 40% of all sports participants are women, yet women’s sports receive only around 4% of all sports media coverage. Increasing the media coverage will help increase the number of role models young girls have.[9]

### Third variable: <a name="3rdvariable"></a>  
### Access to facilities and funding

Women and girls cannot play sport if they cannot get access to the necessary facilities. Too often, sports clubs prioritise male sport when it comes to pitch time so that men and boys get facilities at their preferred times, while women and girls have to make do with less convenient times.[10]      

Funding for female played sports is significantly less than male played sports. As an example in late 2020 it came to light, that €3,000,000 grant funding is paid annually to male GAA inter-county players, whereas only €700,000 is awarded to female GAA inter-county players.   
Separately in soccer, described as “a ground-breaking deal for Irish sport”, in August 2021 the FAI players representing the Republic of Ireland’s senior men’s and women’s football teams are to receive equal match fees with immediate effect. That this was described as ground breaking shows that women do not receive the same funding/pay as their male counterparts. 

In 2014, the House of Commons Committee for Culture, Media and Sport held an inquiry titled ‘Women and Sport’. The inquiry examined the barriers to women’s participation in sport and how to overcome these. The main barriers to female participation in sport and physical activity are as follows:[5]   

-availability of facilities for training and playing sport, for both girls and women, at grassroots and elite levels;  
-finance, including sponsorship and prize money;   
-media coverage of women’s sport;    
-female role models, as elite sportswomen, managers and coaches

## Synthesise/Simulate dataset <a name="synthesiseandsimulate"></a>

### Simulate dataset <a name="simulate"></a>

Based on my research of the variables, I am going to simulate a dataset for 100 females from the main lower age categories, from 10 to 19. This dataset will reflect the current situation in female participation as it stands today. 

```python
# Seeded the random generator to freeze the output and make debugging easier, removed for final submission
np.random.seed(100)

# Take a sample dataset of 100 people between the age of 10 and 19
dfgirls = pd.DataFrame(rng.integers(10,19, 100, endpoint=True), columns=["Age"])
dfgirls
```

```python
# Assign random number of role models to each individual
# A higher probabilities put on the low number of role models to reflect the current environment
dfgirls['Role Models'] = rng.choice(10, size =100, p=[0.22, 0.25, 0.18, 0.12, 0.08, 0.05, 0.04, 0.03, 0.02, 0.01])
dfgirls
```

```python
# Simulate the competitiveness of the sport for each individual 
# As the competitiveness of sport is more prevalent than non competitiveness
# I will assign higher probability to the competitive option, with non-competitive getting a low probability
# Using the numpy random choice with a probability of 65% Competitive, 25% Equally Competitive & Non-Competitive 
# and 10% Non Competitive
types = ['Competitive', 'Equally Competitive & Non-Competitive', 'Non-Competitive']
dfgirls['Competitiveness'] = rng.choice(types,100, p=[0.65, 0.25, 0.10])
dfgirls
```

```python
# Simulate access to facilities and funding versus male counterparts 
# Access to facilities and funding for females is significantly less than their male counterparts
# Therefore I will assign higher probabilities to low funding %
# I am going to assign a zero probability to where female access is higher than male access
# Using the numpy random choice with a probability as follows:
# 75% for <50% access to facilities and funding versus male counterparts
# 25% for 50 to 100% access to facilities and funding versus male counterparts
# 5% for equal access to facilities and funding versus male counterparts
# 0% for where females have more access to facilities and funding versus male counterparts
Access = ['<50%', '50% to 75%','76% to 99%', 'Equal', '+100%']
dfgirls['Access compared to Male counterparts'] = rng.choice(Access,100, p=[0.75, 0.2, 0.05, 0,0])
dfgirls
```

### Categorise dataset for participation rates <a name="synthesise"></a>

Now that we have a realistic dataset, I will group the dataset into participation rate categories:  

**CURRENT RATE** represents female participation at the current levels, i.e. well below the rate of participation of their male counterparts.  
**EXPECTED RATE** represents female participation at the midway point between current levels and level that matches the level of their male counterparts.  
**OPTIMUM RATE** represents female participation that matches the level of their male counterparts.

```python
# Assign participation rate categories to the variables
rmcategory = []
for row in dfgirls['Role Models']:
    if row < 2.0 :    rmcategory.append('Current rate')
    elif row > 6.0:   rmcategory.append('Optimum rate')   
    else:             rmcategory.append('Expected rate')

compcategory = []
for row in dfgirls['Competitiveness']:
    if row == 'Competitive' :   compcategory.append('Current rate')
    elif row == 'Non-Competitive' :   compcategory.append('Optimum rate')    
    else:           compcategory.append('Expected rate')

accesscategory = []
for row in dfgirls['Access compared to Male counterparts']:
    if row == '<50%' :    accesscategory.append('Current rate')
    elif row == '50% to 100%':   accesscategory.append('Expected rate')
    else:           accesscategory.append('Optimum rate')
```

```python
# Add the categorisation to the dataset
dfgirls['Role Model category'] = rmcategory
dfgirls['Competitiveness category'] = compcategory
dfgirls['Access Category'] = accesscategory
print(dfgirls) 
```

```python
# Synthesise the participation rate based on the simulated data
# Define a function that assigns participation rate based on the output of the other functions [11]
def participation_rate(row):
    if row['Role Model category'] == 'Current rate' and row['Competitiveness category'] == 'Current rate' and row['Access Category'] == 'Current rate':
        return 'Current rate'
    if row['Role Model category'] == 'Expected rate' and row['Competitiveness category'] == 'Expected rate' and row['Access Category'] == 'Expected rate':
        return 'Expected rate'
    if row['Role Model category'] == 'Optimum rate' and row['Competitiveness category'] == 'Optimum rate' and row['Access Category'] == 'Optimum rate':
        return 'Optimum rate'
    if row['Role Model category'] == 'Current rate' and row['Competitiveness category'] == 'Current rate':
        return 'Current rate'
    if row['Role Model category'] == 'Current rate' and row['Access Category'] == 'Current rate':
        return 'Current rate'
    if row['Competitiveness category'] == 'Current rate' and row['Access Category'] == 'Current rate':
        return 'Current rate'
    if row['Role Model category'] == 'Expected rate' and row['Competitiveness category'] == 'Expected rate':
        return 'Expected rate'
    if row['Role Model category'] == 'Expected rate' and row['Access Category'] == 'Expected rate':
        return 'Expected rate'
    if row['Competitiveness category'] == 'Expected rate' and row['Access Category'] == 'Expected rate':
        return 'Expected rate'
    if row['Role Model category'] == 'Optimum rate' and row['Competitiveness category'] == 'Optimum rate':
        return 'Optimum rate'
    if row['Role Model category'] == 'Optimum rate' and row['Access Category'] == 'Optimum rate':
        return 'Optimum rate'
    if row['Competitiveness category'] == 'Optimum rate' and row['Access Category'] == 'Optimum rate':
        return 'Optimum rate'
    else:
        return 'Expected rate'

dfgirls['Participation rate'] = dfgirls.apply(lambda row: participation_rate(row), axis=1)
dfgirls.head()
```

Both plots show us that the majority of the females in the dataset participate in sport at the current rate, which is well below the rate of participation of their male counterparts.  
We have about 1/4 at the expected rate representing female participation which is the midway point between current levels and level that equals the level of their male counterparts.   
There is a a minimal number of females participating at the Optimum rate, i.e. participation equal to that of their male counterparts.
