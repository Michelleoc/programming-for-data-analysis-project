```python
dfgirls['Role Models'] = rng.integers(0,10, 100, endpoint=True)
dfgirls
```

```python
types = ['Competitive', 'Non-Competitive']
dfgirls['Competitiveness1'] = rng.choice(types,100, p=[0.8, 0.2])
dfgirls
```

```python
Access = ['<50%', '50% to 100%','Equal', '+100%']
dfgirls['Access'] = rng.choice(Access,100, p=[0.75, 0.2, 0.05, 0])
dfgirls
```
