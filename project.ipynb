```python
rng = np.random.default_rng()
x = rng.choice(2, size =(100), p=[0.63,0.37])
colors = sns.color_palette('pastel')[0:5]
plt.pie(x, colors=colors)
plt.title('Choice with probability assigned', size=14)
plt.xlabel('Number', size=14)
plt.ylabel('Count', size=14)
plt.show()
```

```python
# https://www.rte.ie/brainstorm/2019/0605/1053647-the-challenge-behind-increasing-teenage-girls-physical-activity/

daily_exercise = rng.multinomial(n=100, pvals=[0.623, 0.327])
y = daily_exercise
x = ['Boys', 'Girls']
plt.bar(x,y, color=['blue', 'pink','green'])
plt.xlabel('', size=14)
plt.ylabel('%', size=14)
plt.title('7 year olds meeting daily exercise recommendation', size=14)
plt.show()
```
